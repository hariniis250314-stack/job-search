```python
#Import dependencies
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd
```

```python
title = ["SDE", "Python Developer", "Data Engineer"]  # Job titles
location = "India"  # Job location
start = 0  # Starting point for pagination
```

```python
# Construct the URL for LinkedIn job search and fetch multiple pages
page_jobs = []
num_pages = 5  # Fetch 5 pages to get 100+ results (each page ~25 jobs)

for page in range(num_pages):
    start = page * 25
    list_url = f"https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords={title}&location={location}&start={start}"
    
    # Send a GET request to the URL and store the response
    response = requests.get(list_url)
    
    #Get the HTML, parse the response and find all list items(jobs postings)
    list_data = response.text
    list_soup = BeautifulSoup(list_data, "html.parser")
    jobs = list_soup.find_all("li")
    page_jobs.extend(jobs)
    
    print(f"Page {page + 1}: Found {len(jobs)} jobs")
```

```python
#Create an empty list to store the job postings
id_list = []
```

```python
#Itetrate through job postings to find job ids
for job in page_jobs:
    base_card_div = job.find("div", {"class": "base-card"})
    if base_card_div:
        job_id = base_card_div.get("data-entity-urn").split(":")[3]
        print(job_id)
        id_list.append(job_id)
```

```python
# Initialize an empty list to store job information
job_list = []

# Loop through the list of job IDs and get each URL
for job_id in id_list:
    # Construct the URL for each job using the job ID
    job_url = f"https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}"
    
    # Send a GET request to the job URL and parse the reponse
    job_response = requests.get(job_url)
    print(job_response.status_code)
    job_soup = BeautifulSoup(job_response.text, "html.parser")
    
     # Create a dictionary to store job details
    job_post = {}
    
    # Store the job link
    job_post["job_link"] = f"https://www.linkedin.com/jobs/view/{job_id}"
    
    # Try to extract and store the job title
    try:
        job_post["job_title"] = job_soup.find("h2", {"class":"top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title"}).text.strip()
    except:
        job_post["job_title"] = None
        
    # Try to extract and store the company name
    try:
        job_post["company_name"] = job_soup.find("a", {"class": "topcard__org-name-link topcard__flavor--black-link"}).text.strip()
    except:
        job_post["company_name"] = None
        
    # Try to extract and store the time posted
    try:
        job_post["time_posted"] = job_soup.find("span", {"class": "posted-time-ago__text topcard__flavor--metadata"}).text.strip()
    except:
        job_post["time_posted"] = None
        
    # Try to extract and store the number of applicants
    try:
        job_post["num_applicants"] = job_soup.find("span", {"class": "num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet"}).text.strip()
    except:
        job_post["num_applicants"] = None
    
        
    # Append the job details to the job_list
    job_list.append(job_post)
```

```python
# Create a pandas DataFrame using the list of job dictionaries 'job_list'
jobs_df = pd.DataFrame(job_list)
jobs_df
```

```python
# Install required libraries for vector database and embeddings
import subprocess
import sys

# Install required packages
packages = ['pinecone-client', 'sentence-transformers', 'pypdf2']
for package in packages:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', package])
```

```python
# Import libraries for vector database and embeddings
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import json
import os

# Load the embeddings model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Load the CSV data
df = pd.read_csv('india.csv')

# Create job embeddings and store them
job_embeddings = []
job_data = []

for idx, row in df.iterrows():
    # Combine job title, company, and description for embedding
    job_text = f"{row['job_title']} {row['company_name']}"
    
    # Generate embedding
    embedding = model.encode(job_text)
    job_embeddings.append(embedding)
    
    # Store job data
    job_data.append({
        'id': idx,
        'job_title': row['job_title'],
        'company_name': row['company_name'],
        'job_link': row['job_link'],
        'time_posted': row['time_posted'],
        'num_applicants': row['num_applicants'],
        'embedding': embedding
    })

print(f"Created embeddings for {len(job_data)} jobs")

# Save embeddings to file for later use
np.save('job_embeddings.npy', np.array(job_embeddings))

# Save job metadata
with open('job_data.json', 'w') as f:
    json.dump([{k: (v.tolist() if isinstance(v, np.ndarray) else v) for k, v in job.items()} for job in job_data], f)

print("Embeddings and metadata saved successfully!")
```

```python
# Function to extract text from PDF resume
def extract_text_from_pdf(pdf_path):
    try:
        import PyPDF2
        text = ""
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            for page in pdf_reader.pages:
                text += page.extract_text()
        return text
    except:
        print("Could not read PDF. Please ensure the file is a valid PDF.")
        return None

# Function to match resume with jobs
def find_best_jobs(resume_path, top_k=5):
    """
    Find the best matching jobs for a given resume
    
    Parameters:
    - resume_path: Path to the resume PDF or text file
    - top_k: Number of top matching jobs to return
    
    Returns:
    - List of matching jobs with similarity scores
    """
    
    # Extract resume text
    if resume_path.endswith('.pdf'):
        resume_text = extract_text_from_pdf(resume_path)
    else:
        with open(resume_path, 'r') as f:
            resume_text = f.read()
    
    if resume_text is None:
        return None
    
    # Generate embedding for resume
    resume_embedding = model.encode(resume_text)
    
    # Load job embeddings
    job_embeddings_array = np.load('job_embeddings.npy')
    
    # Calculate similarity scores
    similarities = cosine_similarity([resume_embedding], job_embeddings_array)[0]
    
    # Get top K matches
    top_indices = np.argsort(similarities)[::-1][:top_k]
    
    results = []
    for idx in top_indices:
        results.append({
            'rank': len(results) + 1,
            'job_title': job_data[idx]['job_title'],
            'company_name': job_data[idx]['company_name'],
            'job_link': job_data[idx]['job_link'],
            'similarity_score': float(similarities[idx]),
            'time_posted': job_data[idx]['time_posted'],
            'num_applicants': job_data[idx]['num_applicants']
        })
    
    return results

print("Resume matching function created successfully!")
```

```python
# Example usage - Upload a resume and get job recommendations
# Replace 'your_resume.pdf' with your actual resume file path

resume_file = r"C:\Users\DELL\Downloads\Alex_Chen_Resume.pdf"
  # Change this to your resume file

if os.path.exists(resume_file):
    matching_jobs = find_best_jobs(resume_file, top_k=10)
    
    if matching_jobs:
        print("=" * 80)
        print("TOP MATCHING JOBS FOR YOUR PROFILE")
        print("=" * 80)
        
        for job in matching_jobs:
            print(f"\nRank #{job['rank']}")
            print(f"Job Title: {job['job_title']}")
            print(f"Company: {job['company_name']}")
            print(f"Match Score: {job['similarity_score']:.2%}")
            print(f"Posted: {job['time_posted']}")
            print(f"Applicants: {job['num_applicants']}")
            print(f"Link: {job['job_link']}")
            print("-" * 80)
        
        # Save results to CSV
        results_df = pd.DataFrame(matching_jobs)
        results_df.to_csv('job_recommendations.csv', index=False)
        print("\nResults saved to 'job_recommendations.csv'")
else:
    print(f"Resume file '{resume_file}' not found. Please provide a valid resume file path.")
```
